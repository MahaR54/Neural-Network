# Tokenize Sentences

```python
# Build vocab
tokenized = [sentence.split() for sentence in real_sentences]
vocab = sorted(set(word for sent in tokenized for word in sent))
word2idx = {w: i+1 for i, w in enumerate(vocab)}  # +1 for padding=0
idx2word = {i: w for w, i in word2idx.items()}
vocab_size = len(word2idx) + 1  # for <PAD>=0

# Encode
max_len = max(len(sent) for sent in tokenized)
encoded_data = []
for sent in tokenized:
    idxs = [word2idx[word] for word in sent]
    while len(idxs) < max_len:
        idxs.append(0)  # pad with 0
    encoded_data.append(idxs)

encoded_data = torch.tensor(encoded_data)

print(f"Generated {len(real_sentences)} sentences.")
print(f"Vocabulary size: {vocab_size}")
print(f"Max sequence length: {max_len}")
print(f"Encoded data shape: {encoded_data.shape}")
```

# Generator Code

```python
class Generator(nn.Module):
    def __init__(self, noise_dim, hidden_dim, vocab_size, seq_len):
        super().__init__()
        self.lstm = nn.LSTM(noise_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
        self.seq_len = seq_len
        self.vocab_size = vocab_size

    def forward(self, z):
        h, _ = self.lstm(z)
        logits = self.fc(h)
        probs = F.softmax(logits, dim=-1)
        return probs  # shape: [batch, seq_len, vocab_size]
```

# Discriminator Code

```python
class Discriminator(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, x):
        emb = self.embedding(x)
        _, (h_n, _) = self.lstm(emb)
        out = self.fc(h_n.squeeze(0))
        return torch.sigmoid(out)
```

# Training Loop

```python
def train_text_gan(real_data, vocab_size, seq_len, num_epochs=3000):
    batch_size = 8
    noise_dim = 16
    hidden_dim = 64
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    G = Generator(noise_dim, hidden_dim, vocab_size, seq_len).to(device)
    D = Discriminator(vocab_size, 32, hidden_dim).to(device)

    g_opt = optim.Adam(G.parameters(), lr=0.001)
    d_opt = optim.Adam(D.parameters(), lr=0.001)
    loss_fn = nn.BCELoss()

    real_data = real_data.to(device)

    for epoch in range(num_epochs):
        # === Train Discriminator ===
        D.zero_grad()
        # Sample Real Data,
        # Generate a random sample from given a given population sequence
        real_idx = random.sample(range(real_data.size(0)), batch_size)
        real_batch = real_data[real_idx]

        real_labels = torch.ones(batch_size, 1).to(device)
        fake_labels = torch.zeros(batch_size, 1).to(device)

        # Discriminator on real
        d_real = D(real_batch)   # getting prediction from discriminator
        loss_real = loss_fn(d_real, real_labels)

        # Generate fake
        # random noise
        z = torch.randn(batch_size, seq_len, noise_dim).to(device)
        fake_probs = G(z)
        fake_idx = torch.argmax(fake_probs, dim=-1).detach()  # no grad
        d_fake = D(fake_idx)
        loss_fake = loss_fn(d_fake, fake_labels)

        d_loss = loss_real + loss_fake
        d_loss.backward()
        d_opt.step()

        # === Train Generator ===
        G.zero_grad()
        z = torch.randn(batch_size, seq_len, noise_dim).to(device)
        fake_probs = G(z)
        fake_idx = torch.argmax(fake_probs, dim=-1)
        d_fake = D(fake_idx)
        g_loss = loss_fn(d_fake, real_labels)
        g_loss.backward()
        g_opt.step()

        if epoch % 500 == 0:
            print(f"Epoch {epoch}: D_loss={d_loss.item():.4f}, G_loss={g_loss.item():.4f}")
            print("Sample:")
            sample = fake_idx[0].cpu().tolist()
            print(' '.join(idx2word.get(i, '<PAD>') for i in sample))
            print()
```
