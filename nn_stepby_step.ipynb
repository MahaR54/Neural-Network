```python
def encode(sentence):
    return [word_id[word] for word in sentence.strip().split()]
X = [encode(dataset[d]["question"]) for d in range(0,len(dataset))]
Y = [encode(d["response"]) for d in dataset]
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

X_tensors = [torch.tensor(seq, dtype=torch.long) for seq in X]
Y_tensors = [torch.tensor(seq, dtype=torch.long) for seq in Y]
X_tensors_padded = pad_sequence(X_tensors, batch_first=True)
Y_tensors_padded = pad_sequence(Y_tensors, batch_first=True)
X_tensors_padded.shape
# print(X_tensors_padded.max())
```

```python
import torch.nn as nn

class tGen(nn.Module):
    def __init__(self, vocab_size, emb_dim, hidden_dim):
        super().__init__()
        self.embed = nn.Embedding(vocab_size + 1, emb_dim)
        self.lstm = nn.LSTM(emb_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, vocab_size + 1)

    def forward(self, x):
        x = self.embed(x)
        out, _ = self.lstm(x)
        out = self.fc(out)
        return out
```

```python
import torch.nn.functional as F
import torch

model = tGen(vocab_size=Y_tensors_padded.max().item(), emb_dim=64, hidden_dim=64)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

for epoch in range(100000):
    optimizer.zero_grad()
    output = model.forward(X_tensors_padded)

    target = F.pad(Y_tensors_padded, (0, output.shape[1] - Y_tensors_padded.shape[1]), "constant", 0)

    loss = criterion(output.view(-1, Y_tensors_padded.max().item() + 1), target.view(-1))
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        print(f"Epoch {epoch} Loss: {loss.item():.4f}")
```

```python
def predict_sentence(sentence, model, word_id, id_word):

    import torch
    import re
    def space_special_chars(text):
        text = re.sub(r'([?,.])', r' \1 ', text)
        return re.sub(r'\s+', ' ', text).strip()

    sentence = space_special_chars(sentence)
    encoded = [word_id.get(w, word_id.get('', 0)) for w in sentence.strip().split()]
    input_tensor = torch.tensor([encoded])
    model.eval()
    with torch.no_grad():
        output = model(input_tensor)
        predicted_ids = output.argmax(dim=2)[0].tolist()

    decoded_words = [id_word.get(i, '') for i in predicted_ids if i != 0]
    return ' '.join(decoded_words)
```

```python
import json
import random
from datetime import datetime, timedelta

# Define templates for questions and responses
policy_templates = {
    "Account Management": [
        {
            "questions": [
                "What is the minimum balance for a {} account?",
                "How do I open a {} account with UnityBank?",
                "Can I close my {} account online?",
                "What documents are needed to open a {} account?"
            ],
            "responses": [
                "The minimum balance for a {} account is PKR {}.",
                "Visit a UnityBank branch with your CNIC and {} or apply online.",
                "Yes, log into UnityBank online banking to close your {} account.",
                "You need your CNIC, proof of {}, and a recent bank statement."
            ]
        }
    ],
    "Digital Banking": [
        {
            "questions": [
                "How do I reset my {} banking password?",
                "Is UnityBank's {} banking secure?",
                "Can I update my {} online?",
                "How do I enable {} notifications for my account?"
            ],
            "responses": [
                "Click 'Forgot Password' on the {} banking login page and follow the steps.",
                "Yes, UnityBank uses {} encryption and two-factor authentication for {} banking.",
                "Yes, update your {} in the UnityBank {} banking settings.",
                "Enable {} notifications via UnityBank's mobile app or online banking."
            ]
        }
    ],
    "Lending": [
        {
            "questions": [
                "What documents are required for a {} loan?",
                "What is the maximum amount for a {} loan?",
                "Can I apply for a {} loan online?",
                "What is the interest rate for {} loans?"
            ],
            "responses": [
                "For a {} loan, provide your CNIC, {} proof, and bank statements for {} months.",
                "The maximum {} loan amount depends on your {} and credit history.",
                "Yes, apply for a {} loan through UnityBank's website or mobile app.",
                "The interest rate for {} loans varies; check UnityBank's website for details."
            ]
        }
    ],
    "Fraud Prevention": [
        {
            "questions": [
                "How can I block my {} card?",
                "What is the daily limit for {} transactions?",
                "How does UnityBank protect against {} fraud?",
                "What should I do if I suspect {} fraud?"
            ],
            "responses": [
                "Call UnityBank's helpline at {} or block your {} card via mobile banking.",
                "The daily {} transaction limit is PKR {}.",
                "UnityBank uses {} monitoring and {} authentication to prevent {} fraud.",
                "Report suspected {} fraud to UnityBank's helpline or visit a branch immediately."
            ]
        }
    ],
    "Customer Service": [
        {
            "questions": [
                "What is UnityBank's {} helpline number?",
                "How quickly does UnityBank resolve {} complaints?",
                "Can I get a {} statement by {}?",
                "How do I dispute a {} transaction?"
            ],
            "responses": [
                "Contact UnityBank's {} helpline at {}.",
                "UnityBank resolves {} complaints within {} hours.",
                "Yes, subscribe to {} statements via UnityBank's {}.",
                "Dispute a {} transaction by calling the helpline or visiting a branch with details."
            ]
        }
    ]
}

# Supporting lists for randomization
account_types = ["savings", "current", "student", "business"]
proof_types = ["address", "income", "employment"]
banking_types = ["online", "mobile"]
notification_types = ["SMS", "WhatsApp", "email"]
loan_types = ["personal", "home", "car", "business"]
card_types = ["ATM", "debit", "credit"]
helpline_numbers = ["111-123-456", "111-789-012", "111-456-789"]
timeframes = ["24", "48", "72"]
amounts = [1000, 5000, 10000, 50000, 100000]
security_measures = ["real-time", "AI-based", "automated"]
auth_methods = ["two-factor", "biometric", "multi-factor"]
statement_types = ["e-statement", "account statement", "transaction statement"]
delivery_methods = ["email", "online banking", "mobile app"]

# Function to generate dynamic question-response pairs
def generate_policy(department, template_idx):
    template = policy_templates[department][0]
    question_template = random.choice(template["questions"])
    response_template = random.choice(template["responses"])

    if department == "Account Management":
        account = random.choice(account_types)
        proof = random.choice(proof_types)
        amount = random.choice(amounts)
        question = question_template.format(account)
        response = response_template.format(account, amount if "minimum balance" in response_template else proof, random.randint(3, 12))
    elif department == "Digital Banking":
        banking = random.choice(banking_types)
        notification = random.choice(notification_types)
        question = question_template.format(banking if "banking" in question_template else notification)
        response = response_template.format(banking if "banking" in response_template else notification, random.choice(["256-bit", "end-to-end"]) if "encryption" in response_template else notification)
    elif department == "Lending":
        loan = random.choice(loan_types)
        question = question_template.format(loan)
        response = response_template.format(loan, random.choice(["income", "collateral"]) if "maximum" in response_template else random.choice(proof_types), random.randint(3, 12))
    elif department == "Fraud Prevention":
        card = random.choice(card_types)
        question = question_template.format(card if "card" in question_template else card + " transaction")
        response = response_template.format(random.choice(helpline_numbers) if "helpline" in response_template else card, random.choice(amounts) if "limit" in response_template else random.choice(security_measures), random.choice(auth_methods) if "authentication" in response_template else card)
    elif department == "Customer Service":
        statement = random.choice(statement_types)
        delivery = random.choice(delivery_methods)
        question = question_template.format("customer" if "helpline" in question_template else statement if "statement" in question_template else "customer")
        response = response_template.format("customer" if "helpline" in response_template else statement, random.choice(helpline_numbers) if "helpline" in response_template else delivery, random.choice(timeframes) if "resolve" in response_template else delivery)

    return {"question": question, "response": response}

# Generate 10,000 policies
departments = list(policy_templates.keys())
dataset = []
for i in range(10000):
    department = random.choice(departments)
    policy = generate_policy(department, 0)
    dataset.append(policy)

# Save to JSON
with open("unitybank_policies.json", "w") as f:
    json.dump(dataset, f, indent=2)

print("Dataset with 10,000 UnityBank policies generated and saved as 'unitybank_policies.json'")
```
